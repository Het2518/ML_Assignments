# Custom Covariance Calculation Functions
This section defines basic functions to calculate the mean, mean-centered data, and covariance matrix from scratch, and demonstrates their use on sample vectors.

```python
def calculate_mean_vector(x):
    temp_mean = 0
    for i in range(len(x)):
        temp_mean+=x[i]
    return temp_mean/len(x)

def subtract_mean_from_vector(x,mean):
    x_minus_mean_arr = []
    for i in x:
        x_minus_mean_arr.append(i-mean)
    return x_minus_mean_arr

def calculate_covariance_matrix(x):
    sum_multiply = []
    for i in range(len(x)):
        for j in range(len(x)):
            temp_sum = 0
            for k in range(len(x[i])):
                temp_sum +=(x[i][k]*x[j][k])
            sum_multiply.append(temp_sum/len(x[i]))
    cov_matrix = np.reshape(sum_multiply,(len(x),len(x)))
    return cov_matrix

def get_covariance_from_vectors(x):
    mean_x = []
    for i in x:
        temp = calculate_mean_vector(i)
        mean_x.append(temp)
    x_minus_mean_list = []
    for i in range(len(x)):
        temp2 = subtract_mean_from_vector(x[i],mean_x[i])
        x_minus_mean_list.append(temp2)
    cov_matrix = calculate_covariance_matrix(x_minus_mean_list)
    return cov_matrix
```

```python
moon_values, moon_vectors = np.linalg.eig(moon)
earth_values, earth_vectors = np.linalg.eig(earth)
```

## Eigen Decomposition of Moon and Earth Images
This section computes the eigenvalues and eigenvectors of the Moon and Earth image matrices.

```python
# Reconstruct moon covariance matrix
moon_V = moon_vectors
moon_D = np.diag(moon_values)
moon_V_inv = np.linalg.inv(moon_V)
moon_reconstructed = moon_V @ moon_D @ moon_V_inv

# Reconstruct earth covariance matrix
earth_V = earth_vectors
earth_D = np.diag(earth_values)
earth_V_inv = np.linalg.inv(earth_V)
earth_reconstructed = earth_V @ earth_D @ earth_V_inv
```

```python
moon_close = np.allclose(moon, moon_reconstructed)
earth_close = np.allclose(earth, earth_reconstructed)

print("Moon reconstruction close to original?", moon_close)
print("Earth reconstruction close to original?", earth_close)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Use already loaded moon and earth images
vec1 = moon.flatten().astype(float)
vec2 = earth.flatten().astype(float)

data = np.vstack([vec1, vec2])

mean_data = np.mean(data, axis=1, keepdims=True)
data_centered = data - mean_data

cov_matrix = np.cov(data_centered)

print("Covariance matrix:\n", cov_matrix)

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
D = np.diag(eigenvalues)
V = eigenvectors
V_inv = np.linalg.inv(V)

cov_reconstructed = V @ D @ V_inv

print("Reconstructed covariance matrix:\n", cov_reconstructed)
print("Is reconstruction close?", np.allclose(cov_matrix, cov_reconstructed))

```
